```python
movies_actor_df= pd.read_csv("Resources/actors.tsv", delimiter='\t')
movies_actor_df.head(10)
```

```python
movies_actor_df1=movies_actor_df[["nconst","primaryName","primaryProfession"]]
movies_actor_df1
```

```python
merged_df123= pd.merge(movies_actor_df1,cleaned_df1,on="nconst")
merged_df123
```

```python
df = merged_df123.reindex(columns=['title', 'startYear',"primaryName","category", 'genres','runtimeMinutes','averageRating','numVotes','titleId'])
df
```

```python
df1 = df.drop_duplicates("titleId")
df1
```

```python
df1["genres"].str.split(",")
```

```python
# df1.explode(list("genres"))
df_exploded = df1.explode('genres', ignore_index=True)
 df_exploded
```

```python
temp_df=df_exploded[["titleId","genres"]]
temp_df
```

```python
def encoder(x):
    if x != 0:
        return 1
```

```python
df_pivot = df_exploded.pivot(index='titleId', columns='genres', values="genres").fillna(0)
df_pivot
```

```python
pivot_df=df_pivot.applymap(lambda x: encoder(x)).fillna(0)
pivot_df
```

```python
merged_df14= pd.merge(df_exploded,pivot_df,on="titleId")
merged_df14
```

```python
df3 = merged_df14.drop('genres', axis=1)
df3
```
